### **pd.isin()**

The `pd.isin()` function in Pandas is used to check whether each element in a DataFrame or Series is contained in a sequence of values. Here's an explanation of its functionality:

- **Function**: `pd.isin()`
- **Input**:
  - The values to check for containment can be specified as a list, tuple, NumPy array, or another DataFrame or Series.
- **Output**: Returns a boolean DataFrame or Series indicating whether each element is contained in the specified sequence of values.
- **Usage**:
  - Typically used when you want to filter rows or perform conditional operations based on whether values are present in a given list or array.
- **Example**:
  ```python
  import pandas as pd

  # Create a DataFrame
  df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': ['x', 'y', 'z', 'w']})

  # Check if values in column 'A' are contained in a list
  mask = df['A'].isin([2, 3, 5])

  print(mask)
  ```
  Output:
  ```
  0    False
  1     True
  2     True
  3    False
  Name: A, dtype: bool
  ```
- **Note**:
  - The `pd.isin()` function returns a boolean DataFrame or Series with the same shape as the original DataFrame or Series, where each element indicates whether the corresponding element is contained in the specified sequence of values.
  - You can use the resulting boolean array as a mask to filter rows or perform conditional operations on the DataFrame.
  - If the specified sequence of values contains NaN (missing) values, they will be treated as not contained in the original DataFrame or Series.
  - The function can be applied to a single column (Series) or an entire DataFrame by selecting the desired column(s) to check for containment.

````python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': ['x', 'y', 'z', 'w']})

# Check if values in column 'A' are contained in a list
mask = df['A'].isin([2, 3, 5])

print(mask)
````

### **pd.apply()**

The `pd.apply()` function in Pandas is used to apply a function along an axis of a DataFrame. Here's an explanation of its functionality:

- **Function**: `pd.apply()`
- **Input**:
  - The function to apply can be a NumPy universal function (ufunc), a Python function, or a callable object.
- **Output**: Returns a Series or DataFrame resulting from applying the function along the specified axis.
- **Usage**:
  - Typically used when you want to apply a custom function to each row or column of a DataFrame, or to transform the entire DataFrame using a function.
- **Parameters**:
  - `func`: The function to apply. This can be a function or method to apply to each element of the DataFrame, or a NumPy ufunc to apply element-wise.
  - `axis`: Specifies the axis along which the function should be applied. Default is 0 (apply the function along columns). Use `axis=1` to apply the function along rows.
  - `broadcast`: If set to True, automatically aligns the input function along the specified axis before applying. Default is False.
  - `raw`: If set to True, passes each row or column as a 1D NumPy array to the function. Default is False.
  - `result_type`: Specifies the return type of the function. Options include `'expand'`, `'reduce'`, and `'broadcast'`. Default is None.
- **Example**:
  ```python
  import pandas as pd

  # Create a DataFrame
  df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

  # Define a function to double each element
  def double(x):
      return x * 2

  # Apply the function to each element of column 'A'
  result = df['A'].apply(double)

  print(result)
  ```
  Output:
  ```
  0    2
  1    4
  2    6
  Name: A, dtype: int64
  ```
- **Note**:
  - The function provided to `pd.apply()` is applied element-wise by default, meaning it's applied to each element along the specified axis.
  - The `axis` parameter specifies whether the function should be applied along rows (`axis=1`) or columns (`axis=0`) of the DataFrame.
  - The `pd.apply()` function is flexible and can be used with both simple Python functions and more complex custom functions.
  - If the function returns a Series, the result will be a DataFrame with the same index as the original DataFrame. If it returns a scalar value, the result will be a Series.

````python
# About .apply() function

'''
The .apply() function in pandas is used to apply a given function along the axis of a DataFrame or Series.
It is a versatile and powerful function that allows you to transform data in a DataFrame or Series based on a specified function.

syntax : DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwds)

1. func: The function to apply to each column or row.
2. axis: Specifies whether the function should be applied along rows (axis=0) or columns (axis=1). By default, it's applied to columns (axis=0).
3. raw: If True, the function receives ndarray objects instead of Series.
4. result_type: If specified, it determines the type of the result.
5. args: A tuple that contains additional positional arguments to pass to the function.
6. **kwds: Additional keyword arguments to pass to the function.
'''

#Example - 1
import pandas as pd

# Creating a sample DataFrame
table = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(table)
print(df)
print()

# Applying a function to each column
def square_column(column):
    return column ** 2

result = df.apply(square_column)
print(result)
````

### **pd.groupby()**

The `pd.groupby()` function in Pandas is used to group the rows of a DataFrame based on one or more columns, and then perform aggregation or other operations on these groups. Here's an explanation of its functionality:

- **Function**: `pd.groupby()`
- **Input**:
  - The column or columns to group by.
- **Output**: Returns a GroupBy object, which is a special DataFrameGroupBy object that represents the grouped data.
- **Usage**:
  - Typically used when you want to perform operations on subsets of data based on some categorical variable(s).
- **Parameters**:
  - `by`: Specifies the column or columns to group by. It can be a single column name, a list of column names, or a dictionary mapping column names to group keys.
  - `axis`: Specifies the axis along which to group. Default is 0 (rows).
  - `level`: If the axis is a MultiIndex (hierarchical index), specifies the level(s) to group by.
  - `sort`: Specifies whether to sort the resulting groups by the group keys. Default is True.
  - Other parameters such as `as_index`, `group_keys`, and `squeeze` control various aspects of the group operation.
- **Example**:
  ```python
  import pandas as pd

  # Create a DataFrame
  df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar'],
                     'B': ['one', 'one', 'two', 'two'],
                     'C': [1, 2, 3, 4]})

  # Group the DataFrame by column 'A'
  grouped = df.groupby('A')

  # Calculate the mean of column 'C' for each group
  mean_values = grouped['C'].mean()

  print(mean_values)
  ```
  Output:
  ```
  A
  bar    3.0
  foo    2.0
  Name: C, dtype: float64
  ```
- **Note**:
  - `pd.groupby()` does not actually perform any computation until an aggregation function (e.g., `sum()`, `mean()`, `count()`) is applied to the GroupBy object.
  - The resulting GroupBy object can be used to apply various aggregation functions, such as `sum()`, `mean()`, `count()`, `max()`, `min()`, `agg()`, etc., to compute summary statistics for each group.
  - Grouping can be performed on a single column, multiple columns, or even on levels of a MultiIndex.

````python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar'],
                   'B': ['one', 'one', 'two', 'two'],
                   'C': [1, 2, 3, 4]})

print(df)
````

````python
# Group the DataFrame by column 'A'
grouped = df.groupby('A')

# Calculate the mean of column 'C' for each group
mean_values = grouped['C'].mean()

print(mean_values)
````

````python
# about groupby() function

'''
The groupby() function in pandas is used for grouping rows of a DataFrame based on some criterion.
It is a powerful and flexible tool for data analysis and manipulation. The groupby() operation involves the following steps:

Process of groupby in pandas
1. Splitting : The data is divided into groups based on a specified criterion.
2. Applying a function : A function is applied to each group independently.
3. Combining results : The results are combined into a new data structure.


syntax : DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=False, observed=False, **kwargs)


1. by : This can be a column name, a list of column names, a function, or a dictionary. It determines how the DataFrame should be grouped.
2. axis : Specifies whether the grouping should be done along rows (axis=0) or columns (axis=1).
3. level : If the axis is a MultiIndex, this parameter specifies the level(s) to use for grouping.
4. as_index : If True, the group labels will be used as index. If False, they will be added as new columns.
5. sort : If True, the groups are sorted. If False, the order is the same as the original.
6. group_keys : If True, include group keys in the index (if as_index=True) or as new columns (if as_index=False).
7. squeeze : If True, return a Series if the resulting DataFrame has a single column.
8. observed : If True, it will only show observed values for categorical groupers. This is relevant for categorical data types.


'''

import pandas as pd

table = {
    'Movie': ['Movie1', 'Movie2', 'Movie3', 'Movie4', 'Movie5'],
    'Genre': ['Action', 'Comedy', 'Action', 'Comedy', 'Action'],
    'Rating': [8, 7, 9, 6, 8]
}

df = pd.DataFrame(table)

df
````

````python
# Grouping by the 'Genre' column
grouped = df.groupby('Genre')
print(grouped)
print()

# Applying a function (e.g., calculating the mean) to each group
mean_ratings = grouped['Rating'].mean()
sum_ratings = grouped['Rating'].sum()
max_rating = grouped['Rating'].max()
min_rating = grouped['Rating'].min()
count_ratings = grouped['Rating'].count()


print('average of ratings = ',mean_ratings)
print()
print('sum of ratings = ',sum_ratings)
print()
print('Maximum rating = ',max_rating)
print()
print('Minimum rating = ',min_rating)
print()
print('Count of Ratings = ',count_ratings)
````

### **pd.filter()**

The `pd.filter()` function in Pandas is used to subset or filter a DataFrame based on specific criteria. Here's an explanation of its functionality:

- **Function**: `pd.filter()`
- **Input**:
  - `items`: Specifies the labels (columns or index names) to include or exclude from the result.
  - `like`: Specifies a string or regular expression to match in the column or index names.
  - `regex`: Specifies a regular expression to match in the column or index names.
  - `axis`: Specifies the axis along which to filter. Default is 0 (filter along rows).
- **Output**: Returns a new DataFrame containing only the columns or index levels that match the specified criteria.
- **Usage**:
  - Typically used when you want to select a subset of columns or index levels based on their names or labels.
- **Parameters**:
  - `items`: Specifies the labels (columns or index names) to include or exclude from the result.
  - `like`: Specifies a string or regular expression to match in the column or index names.
  - `regex`: Specifies a regular expression to match in the column or index names.
  - `axis`: Specifies the axis along which to filter. Default is 0 (filter along rows).
- **Example**:
  ```python
  import pandas as pd

  # Create a DataFrame
  df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

  # Filter columns that contain 'A' in the column name
  filtered_df = df.filter(like='A')

  print(filtered_df)
  ```
  Output:
  ```
     A
  0  1
  1  2
  2  3
  ```
- **Note**:
  - The `pd.filter()` function can be used to select columns based on their names or labels using the `items`, `like`, or `regex` parameters.
  - If the `like` parameter is specified, it matches column or index names containing the specified string or regular expression.
  - If the `regex` parameter is specified, it matches column or index names using the specified regular expression pattern.
  - By default, the `pd.filter()` function filters along rows (axis=0). You can specify `axis=1` to filter along columns.

````python
# about .filter() function.


'''
The .filter() function in pandas is used to subset the rows or columns of a DataFrame based on a particular criterion.
It allows you to include or exclude elements from a DataFrame based on a specified condition.

syntax : DataFrame.filter(items=None, like=None, regex=None, axis=None)

1. items: Specifies the columns to include/exclude by name.
2. like: Specifies the columns to include/exclude by substring.
3. regex: Specifies the columns to include/exclude by regular expression.
4. axis: Specifies whether to filter rows (axis=0) or columns (axis=1).

'''


#Example - 1
import pandas as pd

# Creating a sample DataFrame
table = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}

df = pd.DataFrame(table, index=['A', 'B', 'C'])
print(df)
print()

# Filtering rows where 'Age' is greater than 30
filtered_rows = df.filter(like='B', axis=0)

print(filtered_rows)
````

- **Parameters**:
  - `func`: The function to apply. This can be a function or method to apply to each element of the DataFrame or Series, or a NumPy ufunc to apply element-wise.
  - `axis`: Specifies the axis along which the function should be applied. Default is 0 (apply the function along columns). Use `axis=1` to apply the function along rows.
  - `broadcast`: If set to True, automatically aligns the input function along the specified axis before applying. Default is False.
  - `raw`: If set to True, passes each row or column as a 1D NumPy array to the function. Default is False.
  - `result_type`: Specifies the return type of the function. Options include `'expand'`, `'reduce'`, and `'broadcast'`. Default is None.
- **Example**:
  ```python
  import pandas as pd

  # Create a DataFrame
  df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

  # Define a function to double each element
  def double(x):
      return x * 2

  # Apply the function to each element of column 'A'
  result = df['A'].apply(double)

  print(result)
  ```
  Output:
  ```
  0    2
  1    4
  2    6
  Name: A, dtype: int64
  ```
- **Note**:
  - The function provided to `pd.apply()` is applied element-wise by default, meaning it's applied to each element along the specified axis.
  - The `axis` parameter specifies whether the function should be applied along rows (`axis=1`) or columns (`axis=0`) of the DataFrame.
  - The `pd.apply()` function is flexible and can be used with both simple Python functions and more complex custom functions.
  - If the function returns a Series, the result will be a DataFrame with the same index as the original DataFrame. If it returns a scalar value, the result will be a Series.

````python
import pandas as pd

xyz = {
    'roll_no': [1, 2, 1, 3, 1, 3, 3, 3, 2, 2, 1, 2],
    'subject': ['NN', 'DL', 'ML', 'Prob', 'DL', 'ML', 'DL', 'NN', 'NN', 'Prob', 'Prob', 'ML'],
    'marks': [97, 63, 63, 71, 64, 90, 66, 46, 74, 62, 94, 67]
}

df = pd.DataFrame(xyz)
df
````

````python
import pandas as pd

def get_marks(df):
    '''
    INPUT: df -> dataframe

    OUTPUT: result -> dataframe
    '''

    ### STEP 1: group the df using subject and get max marks

    max_marks = df.groupby("subject")["marks"].aggregate("max")

    ### STEP 2: reset the index

    max_marks = max_marks.reset_index()


    ### STEP 3: merge with original df to get roll id

    result = df.merge(max_marks,how="inner",on = ["subject","marks"])[["subject","marks","roll_no"]].sort_values("subject").reset_index(drop=True)

    return result
get_marks(df = pd.DataFrame(xyz))
````
