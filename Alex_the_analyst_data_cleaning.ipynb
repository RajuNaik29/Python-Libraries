```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
df
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
df
```

```python
df.drop(columns = "Not_Useful_Column", inplace = True)
df
```

```python
df['Last_Name'] = df['Last_Name'].str.lstrip("...")
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df['Last_Name'] = df['Last_Name'].str.rstrip("_")
```

```python
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("[^a-zA-Z0-9]", "", regex=True)
```

```python
df
```

```python
df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
str(df['Phone_Number'])
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', "")
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', "")
df
```

```python
df[["Street_address", 'State', 'Zip_code']] = df["Address"].str.split(',', n=2, expand=True)
df
```

```python
df = df.replace("N/a", "")
```

```python
df["Paying Customer"] = df["Paying Customer"].replace("Y", "Yes")
```

```python
df
```

```python
df["Paying Customer"] = df["Paying Customer"].replace("N", "No")
```

```python
df
```

```python
df = df.fillna('')
```

```python
df
```

```python
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Yes':
    df.drop(x, inplace = True)

df
```

```python
df["Do_Not_Contact"] = df['Do_Not_Contact'].replace('N', 'No')
df
```

```python
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace = True)
df
```

```python
df.reset_index(inplace = True)
```

```python
df.drop("index", axis = 1, inplace = True)
```
