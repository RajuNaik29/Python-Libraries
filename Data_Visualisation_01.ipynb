### **About Matplotlib**

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is one of the most widely used plotting libraries and provides a flexible platform for producing high-quality plots for publication, presentations, and exploratory data analysis. Here's an in-depth explanation of Matplotlib:

### Main Components of Matplotlib:

1. **Pyplot Module**: Pyplot is a state-based interface to Matplotlib's plotting library. It provides a MATLAB-like interface for creating plots quickly and easily. Most Matplotlib functions are accessed through the pyplot module.

2. **Figure**: The entire visualization is encapsulated within a figure object. You can think of a figure as a container that holds all the objects representing axes, graphics, text, and labels.

3. **Axes**: An Axes is a specific area of a figure where data is plotted. A figure can contain one or more Axes objects. Each Axes has an x-axis and a y-axis, along with ticks, labels, and other elements that make up the plot.

4. **Artist**: All elements of a Matplotlib plot, including figures, axes, lines, text, etc., are considered Artists. Artists are the objects that compose a figure. They can be arranged in a hierarchy, where the figure is the top-level container, Axes objects are the plotting areas, and individual graphical elements like lines and text are placed within the Axes.

### Types of Plots Supported by Matplotlib:

Matplotlib supports a wide range of plot types, including:

- Line plots
- Scatter plots
- Bar plots
- Histograms
- Box plots
- Pie charts
- Contour plots
- Heatmaps
- 3D plots
- And many more...

### Features of Matplotlib:

1. **Customization**: Matplotlib offers extensive customization options, allowing you to control almost every aspect of the plot, including colors, line styles, markers, fonts, and more.

2. **Publication-Quality Graphics**: Matplotlib produces high-quality plots suitable for publication in scientific journals, presentations, and reports. You can export plots in various formats such as PNG, PDF, SVG, etc.

3. **Support for Multiple Backends**: Matplotlib can generate plots using different backends, including GUI frameworks (e.g., Tkinter, PyQt), web applications (e.g., Flask, Django), and interactive environments (e.g., Jupyter Notebook).

4. **Integration with Pandas and NumPy**: Matplotlib seamlessly integrates with other popular Python libraries like Pandas and NumPy, making it easy to visualize data stored in DataFrame and array formats.

### Usage of Matplotlib:

1. **Interactive Data Exploration**: Matplotlib is commonly used for exploratory data analysis (EDA) to visualize patterns, trends, and relationships in data.

2. **Presentation and Reporting**: Matplotlib is widely used for creating visualizations for presentations, reports, and scientific publications.

3. **Education and Teaching**: Matplotlib is an excellent tool for teaching data visualization concepts in educational settings due to its ease of use and flexibility.

4. **Web Applications and Dashboards**: Matplotlib can be used to create visualizations for web applications and interactive dashboards when combined with frameworks like Flask, Django, or Dash.

Overall, Matplotlib is a powerful and versatile library that enables users to create a wide variety of plots and visualizations, making it an essential tool for data visualization in Python.

### **About matplotlib.pyplot**

Matplotlib is a powerful plotting library for Python widely used for creating static, interactive, and animated visualizations in Python. The `matplotlib.pyplot` module is a collection of functions that allows you to create various types of plots and customize them to suit your needs.

Here's a detailed explanation of `matplotlib.pyplot`:

1. **Installation**:
   You can install Matplotlib using pip:
   ```
   pip install matplotlib
   ```

2. **Importing**:
   After installation, you can import the `matplotlib.pyplot` module in your Python script or notebook:
   ```python
   import matplotlib.pyplot as plt
   ```

3. **Basic Usage**:
   The `pyplot` module provides a MATLAB-like interface for creating and customizing plots. Here are some common functions and their usage:
   - `plt.plot(x, y, linestyle, color, marker, label)`: Create a line plot.
   - `plt.scatter(x, y, color, marker, label)`: Create a scatter plot.
   - `plt.bar(x, height, width, color, label)`: Create a bar plot.
   - `plt.hist(x, bins, color, label)`: Create a histogram.
   - `plt.xlabel(text)`: Set the label for the x-axis.
   - `plt.ylabel(text)`: Set the label for the y-axis.
   - `plt.title(text)`: Set the title of the plot.
   - `plt.legend()`: Show the legend.
   - `plt.show()`: Display the plot.

4. **Figure and Axes**:
   Matplotlib follows a hierarchical structure where a figure contains one or more axes (subplots). You can explicitly create figures and axes using `plt.figure()` and `plt.subplot()`, or you can use the current figure and axes implicitly.

5. **Customization**:
   Matplotlib provides extensive customization options to control various aspects of plots, such as colors, line styles, markers, labels, titles, legends, gridlines, axis limits, and more. You can customize plots using various methods provided by the `pyplot` module or by accessing the attributes of the Figure and Axes objects.

6. **Supported Plot Types**:
   Matplotlib supports a wide range of plot types, including line plots, scatter plots, bar plots, histogram plots, pie charts, box plots, violin plots, heatmaps, contour plots, 3D plots, and more. You can create complex visualizations by combining multiple plot types and customizing them as needed.

7. **Saving Plots**:
   You can save plots as image files (e.g., PNG, JPEG, PDF) using the `plt.savefig()` function.

8. **Interactive Plotting**:
   Matplotlib supports interactive plotting in Jupyter notebooks and other interactive environments using the `%matplotlib notebook` magic command. This allows you to interactively explore and manipulate plots.

9. **Integration**:
   Matplotlib can be integrated with other Python libraries and frameworks, such as NumPy, Pandas, Seaborn, and scikit-learn, to create sophisticated data visualizations and plots.

Overall, `matplotlib.pyplot` provides a flexible and versatile interface for creating high-quality plots and visualizations in Python, making it a popular choice among data scientists, researchers, and developers for data analysis, exploration, and presentation.

````python
# Scatter Plot: Visualizes the relationship between two numerical variables using points.

#syntax : sns.scatterplot(x=None, y=None, hue=None, style=None, size=None, data=None, ...)


import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset('tips')

# Create a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
````

````python
# Bar Plot: Displays the distribution of a numerical variable across categories using bars.


# Syntax = sns.barplot(x='x', y='y', hue=None, data=None, ci='sd', palette='viridis', order=None, hue_order=None)

# Assuming 'data' is a DataFrame containing the data
# x and y are the names of the columns in the DataFrame representing the variables
# hue (optional) is the name of the column used for grouping the data
# data is the DataFrame containing the data
# ci (optional) determines whether to show confidence intervals around the estimate
# palette (optional) is the color palette to use for the bars
# order (optional) is the order in which to display the bars
# hue_order (optional) is the order in which to group the data based on the hue variable


sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot of Total Bill vs Tips')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and a grid of subplots
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)  # Create the first subplot
plt.plot(x, y1)
plt.title('Sin Function')

plt.subplot(1, 2, 2)  # Create the second subplot
plt.plot(x, y2)
plt.title('Cos Function')

plt.tight_layout()  # Optional: Adjust the layout to prevent overlap
plt.show()
````

````python
# Histogram: Displays the distribution of a numerical variable using bins.

# Syntax:sns.histplot(data, x='column_name', bins=None, kde=False, color=None, hue=None, multiple='layer')



plt.figure(figsize=(12, 4))
plt.subplot(2,2,1)
sns.histplot(data = tips, x = 'total_bill', kde = False, bins = 20)
plt.figure(figsize=(12, 4))
plt.subplot(2,2,2)
sns.histplot(data = tips, x = 'total_bill', kde = True, bins = 10)
plt.title('Hist plot of total bill')
plt.subplot(2,2,3)
sns.histplot(data = tips, x = 'total_bill', kde = True, bins = 15, color = 'orange')
plt.title('Hist plot of total bill')

plt.tight_layout()
plt.show()
````

````python
# Box Plot: Displays the distribution of a numerical variable across categories using quartiles.

# syntax : sns.boxplot(x='x_variable', y='y_variable', data=data)

sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
````

````python
# Violin plot : Combines a box plot with a kernel density estimation plot to display the distribution of a numerical variable across categories.

#Syntax : sns.violinplot(x='x_variable', y='y_variable', data=data)

plt.figure(figsize = (10, 5))
plt.subplot(1,2,1)
sns.violinplot(x='day', y='total_bill', data=tips, color = 'orange')
plt.figure(figsize = (10, 5))
plt.subplot(1,2,2)
sns.violinplot(data = tips, x = 'day', y = 'total_bill')

plt.tight_layout()
plt.show()
````

````python
tips
````

````python
# Heatmap : Visualizes the correlation matrix of numerical variables using colors.

# syntax : sns.heatmap(data, cmap='viridis', annot=True, fmt='.2f')


plt.figure(figsize = (10, 5))
plt.subplot(1,2,1)
sns.heatmap(tips.corr(), annot=False, cmap='coolwarm')
plt.figure(figsize = (10, 5))
plt.subplot(1,2,2)
sns.heatmap(tips.corr(), annot = True, cmap = 'viridis', fmt='.1f') #fmt is used as rounding of decimals

plt.tight_layout()
plt.show()
````

````python
# Pair Plot: Displays pairwise relationships between numerical variables in a dataset.
sns.pairplot(tips)
plt.show()
````

````python
# Joint Plot: Visualizes the relationship between two numerical variables using multiple plots, including scatter plots, histograms, and density plots.

# Syntax : sns.jointplot(x='x_variable', y='y_variable', data=data, kind='scatter')

# The value for `kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'], but 'scatterr' was passed.

plt.figure(figsize = (15,10))
plt.subplot(2,3,1)
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.figure(figsize = (15,10))
plt.subplot(2,3,2)
sns.jointplot(data = tips , x = 'total_bill', y = 'tip', kind = 'hist')

plt.figure(figsize = (15,10))
plt.subplot(2,3,3)
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex')

plt.figure(figsize = (15,10))
plt.subplot(2,3,4)
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')

plt.figure(figsize = (15,10))
plt.subplot(2,3,5)
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde')

plt.figure(figsize = (15,10))
plt.subplot(2,3,6)
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'resid')

plt.tight_layout()
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a single figure
plt.figure(figsize=(15, 10))

# Create and display the first joint plot
jointplot1 = sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
jointplot1.fig.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust subplot spacing
plt.show()

# Create and display the second joint plot
jointplot2 = sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist')
jointplot2.fig.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust subplot spacing
plt.show()

# Create and display the third joint plot
jointplot3 = sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
jointplot3.fig.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust subplot spacing
plt.show()

# Create and display the fourth joint plot
jointplot4 = sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
jointplot4.fig.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust subplot spacing
plt.show()

# Create and display the fifth joint plot
jointplot5 = sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
jointplot5.fig.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust subplot spacing
plt.show()

# Create and display the sixth joint plot
jointplot6 = sns.jointplot(data=tips, x='total_bill', y='tip', kind='resid')
jointplot6.fig.subplots_adjust(wspace=0.4, hspace=0.4)  # Adjust subplot spacing
plt.show()
````

### Components of a Matplotlib plot

- Figure: The **overall window** or page that everything is drawn on.
  - You can create multiple independent Figures in Jupyter.
  - If you run the code in terminal, separate windows will pop-up.
- Axes: You can add multiple **Axes** to the Figure, which represents a plot.

- **Axis**: Simply the `x-axis` and `y-axis`

- **Axes**: It is the **area** on which the **data is plotted** with functions such as `plot()`.
    - **x-label**: Name of x-axis
    - **y-label**: Name of y-axis

- **Major ticks**:
  - Subdivides the axis into major units.
  - They appear by default during plotting.
- **Minor ticks**:
  - Subdivides the major tick units.
  - They are by default hidden and can be toggled on.

- **Title**: Title of each plot **(Axes)**

- **Subtitle**: The common title of all the plots.

- **Legend**:
  - Describes the elements in the plot.
  - Blue and Green curves in this case.

These are the major components of a matplotlib plot.

### Boxplot

**What if we want to find the aggregates like median, min, max and percentiles of the data.**

Say I want the typical earnings of a game when it is published.

**What kind of plot can we use here?** Boxplot

**What exactly is a Box plot?**

- A box plot or **box and whiskers plot** shows the **distribution of quantitative data**.
- It facilitates comparisons between
  - attributes
  - across levels

of a categorical attribute.

- The **box** shows the **quartiles** of the dataset.
- The **whiskers** show the **rest of the distribution**.
- Except for points that are determined to be **"outliers"** using a method that is a function of the **inter-quartile range**.

Let's go through the terminology one-by-one.

<!-- #### What's a quartile? -->

<!-- - Box plot shows distribution of numerical data and skewness through displaying the **data percentiles**, called **quartiles**. -->

#### Box plots show the five-number summary of data:
1. Minimum score
2. First (lower) quartile
3. Median
4. Third (upper) quartile
5. Maximum score

##### 1. Minimum Score
- It is the **lowest value**, excluding outliers.
- It is shown at the **end of bottom whisker**.

#### 2. Lower Quartile
- **25% of values** fall below the lower quartile value.
- It is also known as the **first quartile**.

##### 3. Median
- Median marks the **mid-point of the data**.
- It is shown by the **line that divides the box into two parts**.
- **Half the scores are greater than or equal to this value and half are less**.
- It is sometimes known as the **second quartile**.

##### 4. Upper Quartile
- **75% of values** fall below the upper quartile value.
- It is also known as the **third quartile**.

#### Maximum Score
- It is the **highest value**, excluding outliers.
- It is shown at the **end of upper whisker**.

#### Whiskers
- The upper and lower whiskers represent **values outside the middle 50%**.
- That is, the **lower 25% of values** and the **upper 25% of values**.

#### Interquartile Range (or IQR)
- This is the box plot showing the **middle 50% of scores**.
- It is the **range between the 25th and 75th percentile**.

\
**Let's plot a box plot to find the average typical earnings for a game.**

````python
plt.figure(figsize=(15,10))
sns.boxplot(y = data["Global_Sales"])
plt.yticks(fontsize=20)
plt.ylabel('Global Sales (in million dollars)', fontsize=20)
plt.title('Global Sales of video games', fontsize=20)
````

Both code snippets are using the `plt.hist` function in Matplotlib to create a histogram of the 'Year' column from the DataFrame `data`. However, there is a difference in how the bins are specified.

1. In the first code snippet:

```python
counts, bins, img = plt.hist(x=data['Year'])
plt.show()
```

Here, the `plt.hist` function is called with the 'Year' column of the DataFrame as the input (`x=data['Year']`). The function automatically determines the bin edges and computes the histogram. The bin edges and the histogram values are returned in the variables `bins` and `counts`, respectively. The `img` variable likely holds the graphical object for the histogram (though it's not used in the provided code).

2. In the second code snippet:

```python
plt.hist(data['Year'], bins=10)
plt.show()
```

Here, the `plt.hist` function is called with the 'Year' column as the first argument and the number of bins specified as an explicit parameter (`bins=10`). This means the data will be divided into 10 bins.

In summary, the key difference is in how the bins are specified. The first code snippet allows Matplotlib to automatically determine the bin edges, while the second code snippet explicitly sets the number of bins to 10. The choice between these approaches depends on the specific requirements of your analysis and how you want to visualize the data.

````python
# Finding Quater - 1 (Q1)
np.percentile(data['Global_Sales'], 25) # The 25% global sales percent of sales are below 5.58 million.
````

````python
# Finding Median
np.percentile(data['Global_Sales'], 50) # The 50% global sales percent of sales are below 7.53 million.
````

````python
# Finding Quarter - 3 (Q3)
np.percentile(data['Global_Sales'], 75) # The 75% of global sales are below 11.22 million.
````

````python
# IQR - Interquartile Range (Q3 - Q1)
IQR = np.percentile(data['Global_Sales'], 75) - np.percentile(data['Global_Sales'], 25)
IQR
````

````python
# Finding minimum value of above Boxplot

# Formula = Q1 - (1.5 * IQR)

max = np.percentile(data['Global_Sales'], 25) - (1.5*IQR)
max # This is the minimum value of boxplot.
````

````python
# Finding maximum value of above Boxplot.

# Formula = Q3 - (1.5 * IQR)

min = np.percentile(data['Global_Sales'], 75) + (1.5*IQR)
min # This is the maximum value of above boxplot
````
