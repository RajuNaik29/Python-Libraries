### **pd.cut()**

The `pd.cut()` function in Pandas is used to segment and sort data values into bins or intervals. It is particularly useful for converting continuous data into categorical data by discretizing it into bins.

Here's an explanation of its functionality:

- **Function**: `pd.cut()`
- **Input**:
  - `x`: The array-like object containing the data to be segmented into bins.
  - `bins`: Either an integer specifying the number of equal-width bins to create, or a sequence defining the bin edges (inclusive). If the input is an integer, the range of the data will be divided into that many equal-width bins.
  - `labels` (optional): Labels to use for the resulting bins. If not provided, integer labels will be used.
  - Other parameters: Additional parameters for customizing the behavior of the function.
- **Output**: Returns a new categorical Series where each element corresponds to the bin that the original value falls into.
- **Usage**:
  - Typically used to convert continuous data into categorical data by dividing it into bins or intervals.
- **Example**:
  ```python
  import pandas as pd

  # Create a Series with continuous data
  data = pd.Series([1, 5, 10, 15, 20])

  # Segment the data into three bins
  bins = pd.cut(data, bins=3)

  print(bins)
  ```
  Output:
  ```
  0     (0.981, 7.0]
  1     (0.981, 7.0]
  2    (7.0, 13.0]
  3    (13.0, 20.0]
  4    (13.0, 20.0]
  dtype: category
  Categories (3, interval[float64]): [(0.981, 7.0] < (7.0, 13.0] < (13.0, 20.0]]
  ```
- **Note**:
  - The `bins` parameter can be an integer specifying the number of equal-width bins to create or a sequence defining the bin edges (inclusive). If the input is an integer, the range of the data will be divided into that many equal-width bins.
  - The `labels` parameter allows you to specify custom labels for the resulting bins. If not provided, integer labels will be used.
  - The resulting Series is categorical, with each element corresponding to the bin that the original value falls into. This allows for easy grouping and aggregation of data based on the bins.

````python
import pandas as pd

# Create a Series with continuous data
data = pd.Series([1, 2, 3, 4, 5, 5, 6, 9, 11])
data #
````

````python
# Segment the data into two bins
bins = pd.cut(data, bins = 2)

print(bins) # bins that we got are 1. (0.99, 6.0], 2. (6.0, 11.0]
````

````python
# Segment the data into three bins
bins = pd.cut(data, bins = 3)

print(bins) # bins that we got are [(0.99, 4.333] < (4.333, 7.667] < (7.667, 11.0]]
````

````python
# Segment the data into four bins
bins = pd.cut(data, bins = 4)

print(bins) # bins that we got are 1. (0.99,3.5), 2. (3.5, 6.0], 3. (8.5, 11.0]
````

This code snippet utilizes the `.apply()` method in Pandas along with a lambda function to apply a specific operation to each element of a Series or DataFrame. Let's break down each part:

1. `.apply()`: This method applies a function along an axis of the DataFrame or Series. In this case, it's being used with a lambda function.

2. `lambda x: x[-1]`: This lambda function takes each element `x` of the Series and returns the last character of the string. Here, `x[-1]` accesses the last character of the string `x`.

3. `.astype('int')`: This part converts the resulting Series to an integer data type. This is achieved by using the `.astype()` method and specifying the data type as 'int'.

Putting it all together, the code snippet applies a function to each element of a Series. The function extracts the last character of each string element, and then converts the resulting Series to integers.

Here's an example illustrating how this code works:

```python
import pandas as pd

# Create a sample Series
data = pd.Series(['123', '456', '789'])

# Apply the operation to each element
result = data.apply(lambda x: x[-1]).astype('int')

print(result)
```

Output:
```
0    3
1    6
2    9
dtype: int64
```

In this example, each element in the Series is a string containing digits. The code snippet extracts the last character from each string (i.e., the last digit) and converts it to an integer. As a result, we get a new Series with the last digits converted to integers.

````python
import numpy as np
import pandas as pd

dff = {
    'Roll_ID': ['412', np.nan, '456', np.nan, '434', '429', '418'],
    'Name': ['John', 'Mitra', 'Ritz', np.nan, 'Anny', 'Hema', np.nan],
    'Marks': [np.nan, '32', '25', np.nan, '35', '28', '38']
}

df = pd.DataFrame(dff)
df
````

````python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

def preprocess(df):

  """
    Takes df as input, do the preprocessing as asked.
    Return the preprocessed dataframe
  """
  # Drop the rows with all null values
  df.dropna(how = 'all', inplace = True)
  print(df)
  print()

  # Replace the missing values of "Roll_ID" column with 0 and "Name" column with "Anonymous"
  df['Roll_ID'].fillna(0, inplace = True)
  df['Name'].fillna('Anonymous', inplace = True)
  print(df)
  print()

  # Replace the missing values in "Marks" column with the median value of the column
  df.Marks.fillna(df['Marks'].median(),inplace=True)
  print(df)
  print()

  # Change the numerical columns (Roll_ID and Marks) to int datatype in the output
  df['Roll_ID'] = df['Roll_ID'].astype('int')
  df['Marks'] = df['Marks'].astype('int')
  print(df)
  print()

  # Reset the index of the dataframe
  df.reset_index()

  return df

preprocess(df = pd.DataFrame(dff))
````
